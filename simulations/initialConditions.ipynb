```python
import numpy as np
from matplotlib import pyplot as plt
```

# Flow

```python
flow = np.zeros((1000,int(78/38)*1000))
plt.figure(dpi=150)
im = plt.imshow(flow)
plt.colorbar(im)
plt.show()
```

# Temperature

```python
temperature = np.zeros_like(flow)
for i in range(temperature.shape[0]):
    for j in range(temperature.shape[1]):
        cylinderRadius = 4.05/38.4*(temperature.shape[1]/2)
        ycenterCilinder = int(cylinderRadius*(917/1000))
        r = ((ycenterCilinder-i)**2+((temperature.shape[1]/2)-j)**2)**.5 # distance from center of cylinder
        temperature[i,j] = .5*(1.0 - np.tanh((- r + cylinderRadius)/2e1))
plt.figure(dpi=150)
im = plt.imshow(temperature)
plt.colorbar(im)
circle = np.where(np.logical_and(temperature>=0.45, temperature<=0.55))
circle = np.vstack((circle[1],circle[0])).T
# plt.plot(circle[1],circle[0])
plt.show()
```

# Salinity

```python
salinity = np.zeros_like(flow)
for i in range(salinity.shape[0]):
    for j in range(salinity.shape[1]):
        cylinderRadius = 4.05/38.4*(temperature.shape[1]/2)
        ycenterCilinder = int(cylinderRadius*(917/1000))
        r = ((ycenterCilinder-i)**2+((temperature.shape[1]/2)-j)**2)**.5 # distance from center of cylinder
        salinity[i,j] = .5*(1.0 - np.tanh((- r + cylinderRadius)/2e1)) + 1+ np.tanh((i-int(cylinderRadius))/2e1)
plt.figure(dpi=150)
im = plt.imshow(salinity)
plt.colorbar(im)
plt.show()
print('Does salinity have to be nicely defined inside phase = 0?')
```

# Phase

```python
phase = np.zeros_like(flow)
for i in range(phase.shape[0]):
    for j in range(phase.shape[1]):
        cylinderRadius = 4.05/38.4*(temperature.shape[1]/2)
        ycenterCilinder = int(cylinderRadius*(917/1000))
        r = ((ycenterCilinder-i)**2+((temperature.shape[1]/2)-j)**2)**.5 # distance from center of cylinder
        phase[i,j] = .5*(1.0 - np.tanh((r - cylinderRadius)/2e1))
plt.figure(dpi=150)
im = plt.imshow(phase)
plt.colorbar(im)
plt.show()
```
